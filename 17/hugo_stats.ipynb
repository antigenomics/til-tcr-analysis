### PANDAS (TCR+IG)

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
from collections import defaultdict
import seaborn as sns
import itertools

#make dict with sample id on response type
filenames_response = [];filenames_nonresponse = []

with open("annotate_out_TCR/metadata.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        name = l[0]
        resp = l[2]        
        if name != "file_name":
            if resp == "PD":
                filenames_nonresponse.append(name)
            else:
                filenames_response.append(name)  

def pandas_from_vdjtools_annotate(filenames_response, receptor, response):
    panda = collections.defaultdict(list)
    for el1 in filenames_response:
        el = "annotate_out_{}/{}".format(receptor, el1)
        with open(el, "r") as f:
            for line in f.readlines():
                l = line.split("\t")
                if l[0] != "count":
                    if l[12] != "" and l[17] != "" and l[13] != "":
                        if l[12] != "-1":
                            ndn = int(l[12]);v = l[4];d = l[5];j = l[6];insert = int(l[13]);strength = float(l[17]);count = float(l[0]);aa = l[3];el2 = el1.split(".")[0]
                            panda["name"].append(el2);panda["insert_size"].append(insert);panda["strength"].append(strength);panda["ndn_size"].append(ndn)
                            panda["cohort"].append(response);panda["count"].append(count);panda["aa"].append(aa);panda["v"].append(v)
                            panda["d"].append(d);panda["j"].append(j);panda["type"].append(receptor)
    return panda

table_TCR_resp = pd.DataFrame(pandas_from_vdjtools_annotate(filenames_response, "TCR", "response"));table_TCR_nonresp = pd.DataFrame(pandas_from_vdjtools_annotate(filenames_nonresponse, "TCR", "nonresponse"))
table_IG_resp = pd.DataFrame(pandas_from_vdjtools_annotate(filenames_response, "IG", "response"));table_IG_nonresp = pd.DataFrame(pandas_from_vdjtools_annotate(filenames_nonresponse, "IG", "nonresponse"))
frames = [table_TCR_resp,table_TCR_nonresp,table_IG_resp,table_IG_nonresp];result_TCR_IG = pd.concat(frames)
result_TCR_IG
```

### CDF (CUMULATIVE DISTRIBUTION FUNCTION)

### insert_size

```python
result_TCR = result_TCR_IG[result_TCR_IG['type'] == 'TCR']; result_IG = result_TCR_IG[result_TCR_IG['type'] == 'IG']

def CDF_insert_TCR(resp, lol):
    resp_insert = lol.loc[lol['cohort'] == resp].insert_size
    resp_insert2 = resp_insert[resp_insert!=-1]
    resp_insert3 = resp_insert2.sort_values();#add last (largest) value (get unbiased CDF)
    resp_insert3[len(resp_insert3)] = resp_insert3.iloc[-1];#make norm dist massive fron 0 to 1 with len(ser) steps
    cum_dist = np.linspace(0.,1.,len(resp_insert3));resp_insert_cdf = pd.Series(cum_dist, index=resp_insert3)
    return resp_insert_cdf

#IG without add last (largest) value (get unbiased CDF) !!!!
def CDF_insert_IG(resp, lol):
    resp_insert = lol.loc[lol['cohort'] == resp].insert_size
    resp_insert2 = resp_insert[resp_insert!=-1]
    resp_insert3 = resp_insert2.sort_values()
    #add last (largest) value (get unbiased CDF) resp_insert3[len(resp_insert3)] = resp_insert3.iloc[-1]; make norm dist massive fron 0 to 1 with len(ser) steps
    cum_dist = np.linspace(0.,1.,len(resp_insert3));resp_insert_cdf = pd.Series(cum_dist, index=resp_insert3)
    return resp_insert_cdf
        
cdf_ri_TCR =  CDF_insert_TCR("response", result_TCR);cdf_ni_TCR =  CDF_insert_TCR("nonresponse", result_TCR);cdf_ri_IG =  CDF_insert_IG("response", result_IG);cdf_ni_IG =  CDF_insert_IG("nonresponse", result_IG)

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

lin1, = ax1.plot(cdf_ri_TCR, label="response", color="red");lin2, = ax1.plot(cdf_ni_TCR, label="nonresponse", color="green")
ax1.set_title("CDF_insert_Hugo_TCR")
ax1.legend(handles=[lin1, lin2], loc = 'lower right')

lin3, = ax2.plot(cdf_ri_IG, label="response", color="red");lin4, = ax2.plot(cdf_ni_IG, label="nonresponse", color="green")
ax2.set_title("CDF_insert_Hugo_IG")
ax2.legend(handles=[lin3, lin4], loc = 'lower right')

plt.show()
```

### HEATMAP INSERT_SIZE VS STRENGTH

```python
heatmap_TCR_response = result_TCR[(result_TCR.insert_size != -1) & (result_TCR.cohort == "response")][["insert_size","cohort", "strength"]]
heatmap_TCR_nonresponse = result_TCR[(result_TCR.insert_size != -1) & (result_TCR.cohort == "nonresponse")][["insert_size","cohort", "strength"]]
ins_resp = heatmap_TCR_response.insert_size.tolist();stre_resp = heatmap_TCR_response.strength.tolist()
ins_nonresp = heatmap_TCR_nonresponse.insert_size.tolist();stre_nonresp = heatmap_TCR_nonresponse.strength.tolist()

heatmap_IG_response = result_IG[(result_IG.insert_size != -1) & (result_IG.cohort == "response")][["insert_size","cohort", "strength"]]
heatmap_IG_nonresponse = result_IG[(result_IG.insert_size != -1) & (result_IG.cohort == "nonresponse")][["insert_size","cohort", "strength"]]
ins_resp_IG = heatmap_IG_response.insert_size.tolist();stre_resp_IG = heatmap_IG_response.strength.tolist()
ins_nonresp_IG = heatmap_IG_nonresponse.insert_size.tolist();stre_nonresp_IG = heatmap_IG_nonresponse.strength.tolist()

def heatmap(ins, stre):
    d = {}
    for i in range(len(ins)):
        insert = ins[i]
        strength = stre[i]
        key = (insert, strength)
        if key in d.keys():
            d[key] += 1
        else:
            d[key] = 0
    sorted(ins);sorted(stre);n = len(stre);m = len(ins);
    ins = list(set(ins)); stre = list(set(stre)); 
    data = [] 
    for el in ins:
        bla = []
        for k in stre:
            if (el,k) not in d.keys():
                bla.append(0)   
            else:
                bla.append(d[(el,k)])
        data.append(bla)
    return data

data = heatmap(ins_resp, stre_resp);data2 = heatmap(ins_nonresp, stre_nonresp)
data_IG = heatmap(ins_resp_IG, stre_resp_IG);data2_IG = heatmap(ins_nonresp_IG, stre_nonresp_IG)

#heatmap
plt.figure(1, figsize = (10, 8))
plt.subplot(221);sns.heatmap(data, cmap='viridis');plt.title('TCR_response', fontsize=18)
plt.subplot(222);sns.heatmap(data2, cmap='viridis');plt.title('TCR_nonresponse', fontsize=18)
plt.subplot(223);sns.heatmap(data_IG, cmap='viridis');plt.title('IG_response', fontsize=18)
plt.subplot(224);sns.heatmap(data2_IG, cmap='viridis');plt.title('IG_nonresponse', fontsize=18)
plt.tight_layout() 
plt.show()
```

### NUMBER OF READS (IG+TCR)

#### TABLE

```python
d = {}
with open("annotate_out_TCR/metadata.txt") as f:
    for line in f.readlines():
        l = line.split("\t")
        if l[0]!="file_name":
            name = l[0].split(".")[0]
            resp = l[2]
            if l[2] == "PD":
                d[name] = "nonresponse"
            else:
                d[name] = "response"
            
resp2 = defaultdict(list)
for key, val in d.items():
    resp2["name"].append(key)
    resp2["response"].append(val)
resp_on_merge = pd.DataFrame(resp2) 

def RX(meta, receptor):
    hugoTRB = defaultdict(list)
    hugo1 = defaultdict(list)
    #spisok = []
    p = "annotate_out_{}/metadata.txt".format(meta)
    with open(p, "r") as f: 
        for line in f.readlines():
            line1 = line.split("\t")
            if line1[0] != "file_name":
                l = line1[0].split(".")
                l1 = l[0] #id sample
                l3 = "annotate_out_{}/".format(meta)+l1+".txt"
                g = open(l3, "r")
                #spisok.append(l1)
                for st in g.readlines():
                    st1 = st.split("\t")
                    if st1[0] != "count":  
                        if st.find(receptor) != -1:
                            hugo1[l1].append(int(st1[0]))
    #return(hugo1)
    hugoTRB1 = {}
    for key, val in hugo1.items():
        summ = 0
        for el in val:
            summ += el
        hugoTRB1[key] = summ
    for key, val in hugoTRB1.items():
        h = "{}".format(receptor)
        hugoTRB["name"].append(key)
        hugoTRB[h].append(val)   
    hugoTRBpandas = pd.DataFrame(hugoTRB)
    return hugoTRBpandas

TRB = RX("TCR", "TRB");TRA = RX("TCR", "TRA");IGH = RX("IG", "IGH");IGK = RX("IG", "IGK");IGL = RX("IG", "IGL")
AB = TRB.merge(TRA, on = "name", how = "inner").merge(IGH, on = "name", how = "inner").merge(IGL, on = "name", how = "inner").merge(IGK,on = "name", how = "inner")
ABC = AB.merge(resp_on_merge,on = "name", how = "inner")
ABC
```

```python
#dict of lists ==> list of lists ==> common list ===> pandas of unique vals
def common(meta):
    seqaa = defaultdict(list)
    p = "annotate_out_{}/metadata.txt".format(meta)
    with open(p, "r") as f:
        for line in f.readlines():
            line1 = line.split("\t")
            if line1[0] != "file_name":
                l = line1[0].split(".")
                l1 = l[0] #id sample
                l2 = "annotate_out_{}/".format(meta)+l1+".txt"
                g = open(l2, "r")
                for st in g.readlines():
                    st1 = st.split("\t")
                    if st1[0] != "count":
                        seqaa[l1].append(st1[3])
    return seqaa

TCRhugo = common("TCR")
BCRhugo = common("IG") 

spTCRresp = []
#SRR
#pandas with hugo reponse - resp3
srr = []
num = []
for key, val in TCRhugo.items():
        spTCRresp.append(val)
        srr.append(key)
        num.append(len(val))        

u1 = []
num2 = []
for i in range(len(spTCRresp)):
    el = spTCRresp[i]
    new = spTCRresp[:i] + spTCRresp[(i+1):]
    u = []
    newlist = [item for sublist in new for item in sublist]
    for l in el:
        if l not in newlist:
            u.append(l)
    leng = len(u)
    num2.append(leng)
    u1.append(u)
    
#resp
#u1
unique_TCR_hugo_resp = {}
unique_TCR_hugo_resp["name"] = srr
unique_TCR_hugo_resp["unique_clonotypes"] = u1
unique_TCR_hugo_resp["number_of_all_clonotypes"] = num
unique_TCR_hugo_resp["number_of_unique_clonotypes"] = num2
result = pd.DataFrame(unique_TCR_hugo_resp)
result_res = result.merge(resp_on_merge,on = "name", how = "inner")
#result_res
```

### BCR

```python
spBCRresp = []
#SRR
#pandas with hugo reponse - resp3
srr_BCR = []
num_BCR = []
for key, val in BCRhugo.items():
        spBCRresp.append(val)
        srr_BCR.append(key)
        num_BCR.append(len(val))        

u1_BCR = []
num2_BCR = []
for i in range(len(spBCRresp)):
    el = spBCRresp[i]
    new = spBCRresp[:i] + spBCRresp[(i+1):]
    u = []
    newlist = [item for sublist in new for item in sublist]
    for l in el:
        if l not in newlist:
            u.append(l)
    leng = len(u)
    num2_BCR.append(leng)
    u1_BCR.append(u)
    
#resp
#u1
unique_BCR_hugo_resp = {}
unique_BCR_hugo_resp["name"] = srr_BCR
unique_BCR_hugo_resp["unique_clonotypes"] = u1_BCR
unique_BCR_hugo_resp["number_of_all_clonotypes"] = num_BCR
unique_BCR_hugo_resp["number_of_unique_clonotypes"] = num2_BCR
result = pd.DataFrame(unique_BCR_hugo_resp)
result_res = result.merge(resp_on_merge,on = "name", how = "inner")
result_res
```

```python
#fraction of unique clonotypes for TCR
uniq_TCR = {}
uniq_TCR["fraction_of_unique_clonotypes_in_all_samples"] = [(x - y)/x for x, y in zip(num, num2)]
uniq_TCR["name"] = srr
uniq_TCR["type"] = "TCR"
uniq_TCR2 = pd.DataFrame(uniq_TCR)
uniq_TCR2 = uniq_TCR2.merge(resp_on_merge,on = "name", how = "inner")
#fraction of unique clonotypes for IG
uniq_IG = {}
uniq_IG["fraction_of_unique_clonotypes_in_all_samples"] = [(x - y)/x for x, y in zip(num_BCR, num2_BCR)]
uniq_IG["name"] = srr
uniq_IG["type"] = "IG"
uniq_IG2 = pd.DataFrame(uniq_IG)
uniq_IG2 = uniq_IG2.merge(resp_on_merge,on = "name", how = "inner")
#concatenate
frame3 = [uniq_IG2, uniq_TCR2]
unique_reads_all = pd.concat(frame3)
#factorplot
sns.factorplot(kind='box',        # Boxplot
               x='response',        # Y-axis - values for boxplot
               y='fraction_of_unique_clonotypes_in_all_samples',   # X-axis - first factor
               col = 'type',                   
               data=unique_reads_all,        # Dataframe 
               size=4,            # Figure size (x100px)      
               aspect=1.1,        # Width = size * aspect 
               ) 
```
