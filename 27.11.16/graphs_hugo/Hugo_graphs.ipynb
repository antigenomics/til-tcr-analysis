## box-plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import collections
import numpy as np

filenames_PD = [];filenames_PR = [];filenames_CR = [];

with open("metadata.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        name = l[0]
        resp = l[2]        
        if name != "file_name":
            if resp == "PD":
                filenames_PD.append(name)
            elif resp == "PR":
                filenames_PR.append(name)
            else:
                filenames_CR.append(name)      
                
def box_data_insert(filenames_PD):
    insert1 = collections.defaultdict(list)
    sp1 = []
    for el in filenames_PD:
        with open(el, "r") as f:
            insertsize = []
            for line in f.readlines():
                l = line.split("\t")
                if l[0] != "count":
                    if l[12] != "":
                        if l[12] != "-1":
                            b = float(l[12])
                            insertsize.append(b)
            if len(insertsize) != 0:
                avg = sum(insertsize)/len(insertsize)
                sp1.append(avg)
    return sp1 
   #return list of average values of insert_sizes
                
PD_insert = box_data_insert(filenames_PD);PR_insert = box_data_insert(filenames_PR);CR_insert= box_data_insert(filenames_CR);
data_insert  = []
data_insert.append(PD_insert); data_insert.append(PR_insert); data_insert.append(CR_insert);
labels = ["progressive disease", "partial response", "complete response"]

```

```python
fig, ax1 = plt.subplots(figsize=(8, 6))
ax1.set_title('Insert_size_Hugo_boxplot')
#plt.subplots_adjust(left=0.075, right=0.95, top=0.9, bottom=0.25)

bp = plt.boxplot(data_insert, notch=0, sym='+', vert=1, whis=1.5, labels = labels, patch_artist=True)
colors = ['cyan', 'lightgreen', 'pink']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

plt.setp(bp['whiskers'], color='black')
plt.setp(bp['fliers'], color='red', marker='+')

plt.show()
```

```python
def prepare_data(filenames_PD):                
    PD = {}
    ins = set()
    stre = set()
    for el in filenames_PD:
        with open(el, "r") as f:
            for line in f.readlines():
                l = line.split("\t")
                if l[0] != "count":
                    if l[12] != "" and l[13]!= "":
                        if l[12] != "-1":
                            insert = float(l[12])
                            ins.add(insert)
                            strength = float(l[13])
                            stre.add(strength)
                            key = (insert, strength)
                            if key in PD.keys():
                                PD[key] += 1
                            else:
                                PD[key] = 0
    sorted(ins);sorted(stre);n = len(stre);m = len(ins);
    ins = list(ins)
    stre = list(stre)
    data = [] 
    for el in ins:
        bla = []
        for k in stre:
            if (el,k) not in PD.keys():
                bla.append(0)   
            else:
                bla.append(PD[(el,k)])
        data.append(bla)
    return (data, ins, stre)



data, ins, stre = prepare_data(filenames_PD)
data1, ins1, stre1 = prepare_data(filenames_PR)
data2, ins2, stre2 = prepare_data(filenames_CR)

plt.figure(1, figsize = (5, 10))


plt.subplot(311)
plt.pcolor(stre, ins, data)
plt.title('Progressive desease', fontsize=18)

cbar = plt.colorbar()

plt.subplot(312)
plt.pcolor(stre1, ins1, data1)
plt.title('Partial response', fontsize=18)

cbar = plt.colorbar()

plt.subplot(313)
plt.pcolor(stre2, ins2, data2)
plt.title('Complete response', fontsize=18)

cbar = plt.colorbar()

plt.tight_layout() 

plt.show()
```
