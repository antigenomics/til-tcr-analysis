# NDN_SIZE

### box-plot

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.plotly as py
py.sign_in('azulen2', 'n3bcthef3d')
import plotly.graph_objs as go
import collections 

filenames_PD = []
filenames_PR = []
filenames_CR = []


with open("metadata.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        name = l[0]
        resp = l[2]        
        if name != "file_name":
            if resp == "PD":
                filenames_PD.append(name)
            elif resp == "PR":
                filenames_PR.append(name)
            else:
                filenames_CR.append(name)
                

                
insert1 = collections.defaultdict(list)
sp1 = []

for el in filenames_PD:
    with open(el, "r") as f:
        insertsize = []
        for line in f.readlines():
            l = line.split("\t")
            if l[0] != "count":
                if l[11] != "":
                    b = float(l[11])
                    insertsize.append(b)
        if len(insertsize) != 0:
            avg = sum(insertsize)/len(insertsize)
            sp1.append(avg)
            #print (avg)
            
insert1["PD"] = sp1
            
insert2 = collections.defaultdict(list)
sp2 = []

for el in filenames_PR:
    with open(el, "r") as f:
        insertsize = []
        for line in f.readlines():
            l = line.split("\t")
            if l[0] != "count":
                if l[11] != "":
                    b = float(l[11])
                    insertsize.append(b)
        if len(insertsize) != 0:
            avg = sum(insertsize)/len(insertsize)
            sp2.append(avg)
            
            
insert2["PR"] = sp2
            
            
            
insert3 = collections.defaultdict(list)
sp3 = []

for el in filenames_CR:
    with open(el, "r") as f:
        insertsize = []
        for line in f.readlines():
            l = line.split("\t")
            if l[0] != "count":
                if l[11] != "":
                    b = float(l[11])
                    insertsize.append(b)
        if len(insertsize) != 0:
            avg = sum(insertsize)/len(insertsize)
            sp3.append(avg)
            
            
insert3["CR"] = sp3

 
insert1.update(insert2)

insert1.update(insert3)

x_PD = []
y_PD = []

x_PR = []
y_PR = []

x_CR = []
y_CR = []

for key, val in insert1.items():
    if key == "PD":
        x_PD.append(key)
        y_PD.append(val)
    elif key == "PR":
        x_PR.append(key)
        y_PR.append(val)
    else: 
        x_CR.append(key)
        y_CR.append(val)
```

```python
traces = []

for xd, yd in zip(x_PD, y_PD):
        traces.append(go.Box(
            y=yd,
            name="progressive disease",
            #boxpoints='all',
            jitter=0.5,
            whiskerwidth=0.2,   
            marker=dict(
                color = 'rgb(214, 12, 140)',
            ),
            line=dict(width=1),
        ))        

for xd, yd in zip(x_PR, y_PR):
        traces.append(go.Box(
            y=yd,
            name="partial response",
            #boxpoints='all',
            jitter=0.5,
            whiskerwidth=0.2,   
            marker=dict(
                color = 'rgb(0, 128, 128)',
            ),
            line=dict(width=1),
        ))  

for xd, yd in zip(x_CR, y_CR):
        traces.append(go.Box(
            y=yd,
            name="complete response",
            #boxpoints='all',
            jitter=0.5,
            whiskerwidth=0.2,   
            marker=dict(
                color = 'rgba(127, 96, 0, 0.5)',
            ),
            line=dict(width=1),
        ))  
        
layout = go.Layout(title = 'ndn_size_HUGO', showlegend=False)

fig = go.Figure(data=traces, layout = layout)
py.iplot(fig)  
```

```python
py.image.save_as(fig, filename='ndn_HUGO_box_plot.png')
```

### Histogram

```python
import plotly.plotly as py
import numpy as np
import plotly.graph_objs as go
import numpy as np

PD1 = []
PR1 = []
CR1 = []

for el in filenames_PD:
    with open(el, "r") as f:
        insertsize = []
        for line in f.readlines():
            l = line.split("\t")
            if l[0] != "count":
                if l[11] != "":
                    b = float(l[11])
                    PD1.append(b)
                

PD = []
sum_PD = sum(PD1)
for el in PD1:
    PD.append(el/sum_PD)

for el in filenames_CR:
    with open(el, "r") as f:
        insertsize = []
        for line in f.readlines():
            l = line.split("\t")
            if l[0] != "count":
                if l[11] != "":
                    b = float(l[11])
                    CR1.append(b) 
                    
CR = []
sum_CR = sum(CR1)
for el in CR1:
    CR.append(el/sum_CR)      
    
for el in filenames_PR:
    with open(el, "r") as f:
        insertsize = []
        for line in f.readlines():
            l = line.split("\t")
            if l[0] != "count":
                if l[11] != "":
                    b = float(l[11])
                    PR1.append(b)  
                    
PR = []
sum_PR = sum(PR1)
for el in PR1:
    PR.append(el/sum_PR)
```

```python
trace1 = go.Histogram(
    name = 'complete response',
    x=CR,
    opacity=0.75
)

trace2 = go.Histogram(
    name = 'partial response',
    x=PR,
    opacity=0.75
)

trace3 = go.Histogram(
    name = 'progressive disease',
    x=PD,
    opacity=0.75
)


data = [trace1, trace2, trace3]
layout = go.Layout(
    barmode='overlay'
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

```python
py.image.save_as(fig, filename='ndn_HUGO_hist.png')
```

### Kruskall wallis

```python
from scipy import stats
stats.kruskal(PR1, PD1, CR1)
```

p_val < stat => H0 reject (medians are not equal)

### 1 way ANOVA

```python
import scipy.stats as stats
stats.f_oneway(PR1, PD1, CR1)
```

pval < stat => H0 reject (means are not equal)

### CDF

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array(PR1) 
num_bins = 20
counts, bin_edges = np.histogram(a, bins=num_bins, normed=True)
cdf = np.cumsum(counts)
line_med, = plt.plot(bin_edges[1:], cdf, label="partial response", color = "blue")

a = np.array(PD1) 
num_bins = 20
counts, bin_edges = np.histogram(a, bins=num_bins, normed=True)
cdf = np.cumsum(counts)
line_up, = plt.plot(bin_edges[1:], cdf,  label="progressive disease", color="green")

a = np.array(CR1) 
num_bins = 20
counts, bin_edges = np.histogram(a, bins=num_bins, normed=True)
cdf = np.cumsum(counts)
line_down, = plt.plot(bin_edges[1:], cdf,  label="complete response", color="red")

plt.title("CDF_ndn_hugo")

plt.legend(handles=[line_up, line_down, line_med])

plt.show()
```
