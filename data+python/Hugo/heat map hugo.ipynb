```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import summer

filenames_PD = []
filenames_PR = []
filenames_CR = []

with open("metadata.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        name = l[0]
        resp = l[2]        
        if name != "file_name":
            if resp == "PD":
                filenames_PD.append(name)
            elif resp == "PR":
                filenames_PR.append(name)
            else:
                filenames_CR.append(name)   
                            
def prepare_data(filenames_PD):                
    PD = {}
    ins = set()
    stre = set()
    for el in filenames_PD:
        with open(el, "r") as f:
            for line in f.readlines():
                l = line.split("\t")
                if l[0] != "count":
                    if l[12] != "" and l[13]!= "":
                        if l[12] != "-1":
                            insert = float(l[12])
                            ins.add(insert)
                            strength = float(l[13])
                            stre.add(strength)
                            key = (insert, strength)
                            if key in PD.keys():
                                PD[key] += 1
                            else:
                                PD[key] = 0
    sorted(ins);sorted(stre);n = len(stre);m = len(ins);
    ins = list(ins)
    stre = list(stre)
    data = [] 
    for el in ins:
        bla = []
        for k in stre:
            if (el,k) not in PD.keys():
                bla.append(0)   
            else:
                bla.append(PD[(el,k)])
        data.append(bla)
    return (data, ins, stre)



data, ins, stre = prepare_data(filenames_PD)
data1, ins1, stre1 = prepare_data(filenames_PR)
data2, ins2, stre2 = prepare_data(filenames_CR)
```

```python
plt.figure(1, figsize = (15, 15))


plt.subplot(311)
plt.pcolor(stre, ins, data)
plt.title('Progressive desease', fontsize=18)

cbar = plt.colorbar()

plt.subplot(312)
plt.pcolor(stre1, ins1, data1)
plt.title('Partial response', fontsize=18)

cbar = plt.colorbar()

plt.subplot(313)
plt.pcolor(stre2, ins2, data2)
plt.title('Complete response', fontsize=18)

cbar = plt.colorbar()

plt.tight_layout() 

plt.show()

```
