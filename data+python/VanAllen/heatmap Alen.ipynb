```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import summer

filenames_resp = []
filenames_nonresp = []


with open("metadata.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        name = l[0]
        resp = l[2]        
        if name != "file_name":
            if resp == "response":
                filenames_resp.append(name)
            if resp == "nonresponse":
                filenames_nonresp.append(name)

def prepare_data(filenames_PD):                
    PD = {}
    ins = set()
    stre = set()
    for el in filenames_PD:
        with open(el, "r") as f:
            for line in f.readlines():
                l = line.split("\t")
                if l[0] != "count":
                    if l[12] != "" and l[13]!= "":
                        if l[12] != "-1":
                            insert = float(l[12])
                            ins.add(insert)
                            strength = float(l[13])
                            stre.add(strength)
                            key = (insert, strength)
                            if key in PD.keys():
                                PD[key] += 1
                            else:
                                PD[key] = 0
    sorted(ins);sorted(stre);n = len(stre);m = len(ins);
    ins = list(ins)
    stre = list(stre)
    data = [] 
    for el in ins:
        bla = []
        for k in stre:
            if (el,k) not in PD.keys():
                bla.append(0)   
            else:
                bla.append(PD[(el,k)])
        data.append(bla)
    return (data, ins, stre)



data, ins, stre = prepare_data(filenames_resp)
data1, ins1, stre1 = prepare_data(filenames_nonresp)
```

```python
plt.figure(1, figsize = (15, 15))


plt.subplot(211)
plt.pcolor(stre, ins, data)
plt.title('response', fontsize=18)

cbar = plt.colorbar()

plt.subplot(212)
plt.pcolor(stre1, ins1, data1)
plt.title('nonresponse', fontsize=18)

cbar = plt.colorbar()

plt.tight_layout() 

plt.show()
```
