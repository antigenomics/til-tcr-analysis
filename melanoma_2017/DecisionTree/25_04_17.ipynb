### JS DISTANCE BETWEEN NORMA AND MELANOMA V GENES

#### TRB

```python
patients_TRB = set()
variants_v_TRB = set()
d_TRB = {}
with open("melanoma_good.txt", "r") as f:
    for line in f.readlines():
        if line.find("patient_id") == -1:
            l = line.split("\t")
            pat = l[1]
            #print (pat)
            #cdr = l[6]
            if l[2] == 'TRB':
                patients_TRB.add(pat)
                v = l[9]
                variants_v_TRB.add(v)
                key = (v, pat)
                if key in d_TRB.keys():
                    d_TRB[key] += 1
                else:
                    d_TRB[key] = 1
                    
data_TRB = []

for el in variants_v_TRB:
    bla = []
    for k in patients_TRB:
        if (el,k) not in d_TRB.keys():
            bla.append(0)
        else:
            bla.append(d_TRB[(el,k)])
    data_TRB.append(bla)

#data matrix to DataFrame 
p2 = defaultdict(list)
for (a,b) in zip(data_TRB, variants_v_TRB):
    p2[b] = a

TRB_mel = pd.DataFrame(p2, index = patients_TRB)

TRB_mel2 = TRB_mel.div(TRB_mel.sum(axis=1), axis=0)
TRB_mel2['key1'] = pd.Series(list(patients_TRB), index=TRB_mel2.index)
TRB_mel3 = TRB_mel2.dropna(thresh=2) #drop null row if exists
TRB_mel3[0:3]

TRB_mel4 = TRB_mel3.ix[:, TRB_mel3.columns != 'key1']
TRB_mel4[0:2]
```

```python
#jensen-shannon metric
def JSD(P, Q):
    _P = P / norm(P, ord=1)
    _Q = Q / norm(Q, ord=1)
    _M = 0.5 * (_P + _Q)
    return 0.5 * (entropy(_P, _M) + entropy(_Q, _M))


#distance vector
def dist2(a, b):
    items = a.shape[0] #(466, 279) --> 466
    out = np.ones(items) #
    for j in range(items):
        out[j] = JSD(a[j], b)
    return out

#main 
def func_JSD_TRB (filename, TRB_mel4, num):
    a = pd.read_csv(filename, sep="\t")  
    #make counter for trb occurence 
    b = dict(Counter(a['v'].tolist()))
    
    #find missing V
    keys_b = b.keys()
    keys_mel = TRB_mel4.columns.tolist()
    intersect1 = list(set(keys_mel) - set(keys_b))
    intersect2 = list(set(keys_b) - set(keys_mel))
    
    #add 0 for missing V
    for el in intersect1:
        b[el] = 0
        
    #add 0 columns to melanoma
    for el1 in intersect2:
        TRB_mel4[el1] = 0
        
    #make norma dataframe    
    dnew = defaultdict(list)
    for k,v in b.items():
        dnew['V'].append(k)
        dnew['count'].append(v)
    e = pd.DataFrame(dnew) 
    
    #order from norma to melanomva
    norma_order = e['V'].tolist()
    
    #ordering melanoma
    TRB_mel4 = TRB_mel4[norma_order]
    
    #normalise norma
    e_norm = e['count'].div(e['count'].sum()).tolist()
    
    #melanoma to matrix
    TRB_mel4_matrix = TRB_mel4.as_matrix()
    
    #JSdist for matrix and vector
    distTRBvsNORMA1 = dist2(TRB_mel4_matrix, e_norm)
    
    #make dataframe
    colname = 'norma_sample'+num
    pd_distTRB = pd.DataFrame({'patients': TRB_mel4.index.tolist(), colname: list(distTRBvsNORMA1)})
    
    return pd_distTRB 
```

```python
a1 = func_JSD_TRB("norma/A4-i194.txt", TRB_mel4, '1')
a2 = func_JSD_TRB("norma/A2-i136.txt", TRB_mel4, '2')
a3 = func_JSD_TRB("norma/A4-i122.txt", TRB_mel4, '3')
a4 = func_JSD_TRB("norma/A3-i150.txt", TRB_mel4, '4')
a5 = func_JSD_TRB("norma/A3-i102.txt", TRB_mel4, '5')
a6 = func_JSD_TRB("norma/A2-i131.txt", TRB_mel4, '6')

#merge all 
a_all = [a1, a2, a3, a4, a5, a6]
a_final = reduce(lambda left,right: pd.merge(left,right,on='patients', how='inner'), a_all)

```

```python
def func_JSD_IGH(filename, TRB_mel4, num):
    a = pd.read_csv(filename, sep="\t")  
    #make counter for trb occurence 
    #IGHV4-38-2*01
    b = dict(Counter([x.split("*")[0] for x in a['v'].tolist()]))
    
    #find missing V
    keys_b = b.keys()
    keys_mel = TRB_mel4.columns.tolist()
    intersect1 = list(set(keys_mel) - set(keys_b))
    intersect2 = list(set(keys_b) - set(keys_mel))
    
    #add 0 for missing V in norma
    for el in intersect1:
        b[el] = 0
        
    #add 0 columns for missing in melanoma
    for el1 in intersect2:
        TRB_mel4[el1] = 0
        
    #make norma dataframe    
    dnew = defaultdict(list)
    for k,v in b.items():
        dnew['V'].append(k)
        dnew['count'].append(v)
    e = pd.DataFrame(dnew) 
    
    #order from norma to melanomva
    norma_order = e['V'].tolist()
    
    #ordering melanoma
    TRB_mel4 = TRB_mel4[norma_order]
    
    #normalise norma
    e_norm = e['count'].div(e['count'].sum()).tolist()
    
    #melanoma to matrix
    TRB_mel4_matrix = TRB_mel4.as_matrix()
    
    #JSdist for matrix and vector
    distTRBvsNORMA1 = dist2(TRB_mel4_matrix, e_norm)
    
    #make dataframe
    colname = 'norma_sample'+num
    pd_distTRB = pd.DataFrame({'patients': TRB_mel4.index.tolist(), colname: list(distTRBvsNORMA1)})
    
    return pd_distTRB 
```
