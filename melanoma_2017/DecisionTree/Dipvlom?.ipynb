### Survival and Stage data

```python
survival = pd.read_csv("survival.txt", sep='\t', header= None)
survival.columns = ["name", "surv", "1", "2", "3", "4"]
survival[0:5]

#open TCGA to SKCM ids
ids = pd.read_csv("TCGA_to_SKCM.csv", sep='\t')
ids[0:5]

#merge survival with ids
surv_data = survival.merge(ids, on="name", how="inner")
surv_data[0:5]

#delete None and <0, make column type numeric
s = surv_data.loc[surv_data['surv']!= "None"]
s[['surv', '2']] = s[['surv', '2']].apply(pd.to_numeric)
s = s.loc[s['surv'] >= 0]

s[0:5]

set(s['3'].tolist())

def alert(s):
    if (s["3"] == 'Ia') | (s["3"] == 'Ib') | (s["3"] == 'I'):
        return 'I'
    elif (s["3"] == 'II') | (s["3"] == 'IIb') | (s["3"] == 'IIa') | (s["3"] == 'IIc') | (s["3"] == 'I/II nos'):
        return 'II'
    elif (s["3"] == 'III') | (s["3"] == 'IIIb') | (s["3"] == 'IIIa') | (s["3"] == 'IIIc'):
        return 'III'
    elif (s["3"] == 'Iv'):
        return 'Iv'
    else:
        return '0'
        
s['class_status'] = s.apply(alert, axis=1)
s[0:5]
```

### Clonality.gini

```python
def gini(list_of_values):
    sorted_list = sorted(list_of_values)
    height, area = 0, 0
    for value in sorted_list:
        height += value
        area += height - value / 2.
        #print (height, area)
    fair_area = height * len(list_of_values) / 2.
    #print (fair_area)
    return (fair_area - area) / fair_area

ids = pd.read_csv("TCGA_to_SKCM.csv", sep='\t')
ids.drop('Unnamed: 0', axis=1, inplace=True) 
ids[0:5]

patients_all = set()
number_of_reads_all = set()
d_all = defaultdict(list)


with open("melanoma.txt", "r") as f:
    for line in f.readlines():
        if line.find("patient_id") == -1:
            l = line.split("\t")
            pat = l[0]
            patients_all.add(pat)
            read = float(l[3])
            d_all[pat].append(read)
 
                
d2_all = defaultdict(list)
for k, v in d_all.items():
    val = gini(v)
    d2_all["name"].append(k)
    d2_all["gini"].append(val)
    
gini_all = pd.DataFrame(d2_all)
gini_all[0:5]  
```

```python
gini_IGreads = number_B_TCGA.merge(gini_all, on='name', how='inner')
gini_IGreads.boxplot(column='gini',by='class_status',figsize=(10,5))
```

### Vusage vs stage clustermap

```python
patients = set()
variants_v = set()
d = {}

with open("melanoma.txt", "r") as f:
    for line in f.readlines():
        if line.find("patient_id") == -1:
            l = line.split("\t")
            pat = l[0]
            #cdr = l[6]
            patients.add(pat)
            v = l[8].split(",")
            for el in v:
                variants_v.add(el)
                key = (el, pat)
                if key in d.keys():
                    d[key] += 1
                else:
                    d[key] = 1
                    
data = []
for el in variants_v:
    bla = []
    for k in patients:
        if (el,k) not in d.keys():
            bla.append(0)
        else:
            bla.append(d[(el,k)])
    data.append(bla)

#data matrix to DataFrame 
p2 = defaultdict(list)
for (a,b) in zip(data, variants_v):
    p2[b] = a

df2 = pd.DataFrame(p2, index = patients)
df2[0:3]

df3 = df2.div(df2.sum(axis=1), axis=0)
df3['key1'] = pd.Series(list(patients), index=df3.index)
#df4 = df3.dropna(thresh=2) #drop null row if exists
df3[0:3]

import math
from scipy.stats import entropy
from numpy.linalg import norm
import numpy as np

a = df3.ix[:, 0:-1].as_matrix()
a

#jensen-shannon metric
def JSD(P, Q):
    _P = P / norm(P, ord=1)
    _Q = Q / norm(Q, ord=1)
    _M = 0.5 * (_P + _Q)
    return 0.5 * (entropy(_P, _M) + entropy(_Q, _M))

#distance matrix
def dist(a):
    items = a.shape[0] #(466, 279) --> 466
    out = np.ones((items,items)) #
    for j in range(items):
        for k in range(j+1, items):
            out[j, k] = JSD(a[j], a[k])
            out[k, j] = out[j, k]
    return out

b = dist(a)


titles = df3.index.tolist()
row_dist = pd.DataFrame(b, index = titles)

column_indices =  [x for x in range(466)]
old_names = row_dist.columns[column_indices]
new_names = titles
row_dist.rename(columns=dict(zip(old_names, new_names)), inplace=True) #replace int indexes to SKCM_id

row_dist[0:3]

df = df3.ix[:, 0:-1]

#color stage pandas
s1 = s[['name', 'class_status']]
def func(s):
    if s['class_status'] == '0':
        return 'red'
    elif s['class_status'] == 'I':
        return 'green'
    elif s['class_status'] == 'II':
        return 'yellow'
    elif s['class_status'] == 'III':
        return 'blue'
    else:
        return 'orange'

s1['status_color'] = s1.apply(func, axis=1)
s2 = s1[['name', 'status_color']]

s2.index = list(s2['name'])
color_row = s2.drop(s2.columns[[0]], axis=1)

#color legend
color_dict = {'0':'red', 'I':'green', 'II':'yellow', 'III':'blue', 'IV':'orange'}
color_dict

#linkage
import scipy.spatial as sp, scipy.cluster.hierarchy as hc
linkage = hc.linkage(row_dist, method='ward')
```

```python
grid = sns.clustermap(df, row_linkage=linkage, figsize = (50, 50), row_colors=color_row, col_cluster=False)

#
grid.cax.set_visible(False)

#SKCM labels to horizontal
plt.setp(grid.ax_heatmap.get_yticklabels(), rotation=0)

#
grid.ax_heatmap.set_ylabel("");

#legend
for label in color_dict.keys():
    grid.ax_col_dendrogram.bar(0, 0, color=color_dict[label],
                            label=label, linewidth=0)
    
grid.ax_col_dendrogram.legend(loc="center", ncol=6)

grid.savefig("output4.png")
```
