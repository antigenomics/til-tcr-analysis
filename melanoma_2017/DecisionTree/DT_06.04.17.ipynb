```python
%matplotlib inline
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import collections
from collections import defaultdict
import seaborn as sns
import itertools
import scipy.cluster.hierarchy as hac
import matplotlib.pyplot as plt
```

### data

```python
#open survival.txt
survival = pd.read_csv("survival.txt", sep='\t', header= None)
survival.columns = ["name", "surv", "1", "2", "3", "4"]
survival[0:5]

#open TCGA to SKCM ids
ids = pd.read_csv("TCGA_to_SKCM.csv", sep='\t')
ids[0:5]

#merge survival with ids
surv_data = survival.merge(ids, on="name", how="inner")
surv_data[0:5]

#delete None and <0, make column type numeric
s = surv_data.loc[surv_data['surv']!= "None"]
s[['surv', '2']] = s[['surv', '2']].apply(pd.to_numeric)
s = s.loc[s['surv'] >= 0]

s = s[s['3'] != 'not reported']

s[0:5]
```

### Censorship

```python
def alert(c):
  if c['1'] == 'dead':
    return 1
  else:
    return 0

s['status'] = s.apply(alert, axis=1)
s[0:5]
```

```python
T = s['surv']
E = s['status']
from lifelines import KaplanMeierFitter
kmf = KaplanMeierFitter()
kmf.fit(T, event_observed=E) 
kmf.survival_function_
kmf.plot()
```

```python
kmf.median_
```

```python
stage_type = sorted(s['3'].unique())
T = s['surv']
C = s['status']
```

```python
for i,stage1 in enumerate(stage_type):
    ax = plt.subplot(5,3,i+1)
    ix = s['3'] == stage1
    kmf.fit( T[ix], C[ix], label=stage1)
    kmf.plot(ax=ax, legend=False, figsize = (10,18))
    plt.title(stage1)
    #plt.xlim(0,50)
    if i==0:
        plt.ylabel('probability')
plt.tight_layout()
```

```python
def alert(s):
    if (s["3"] == 'Ia') | (s["3"] == 'Ib') | (s["3"] == 'I') | (s["3"] == 'I/II nos'):
        return 'I'
    elif (s["3"] == 'II') | (s["3"] == 'IIb') | (s["3"] == 'IIa') | (s["3"] == 'IIc'):
        return 'II'
    elif (s["3"] == 'III') | (s["3"] == 'IIIb') | (s["3"] == 'IIIa') | (s["3"] == 'IIIc'):
        return 'III'
    elif (s["3"] == 'Iv'):
        return 'Iv'
    else:
        return '0'
        
s['class_status'] = s.apply(alert, axis=1)
s[0:4]
```

```python
kmf = KaplanMeierFitter()
T1 = s[s['class_status'] == 'I'].surv
C1 = s[s['class_status'] == 'I'].status

T2 = s[s['class_status'] == 'II'].surv
C2 = s[s['class_status'] == 'II'].status

T3 = s[s['class_status'] == 'III'].surv
C3 = s[s['class_status'] == 'III'].status

T4 = s[s['class_status'] == 'Iv'].surv
C4 = s[s['class_status'] == 'Iv'].status

kmf.fit(T1, C1, label='I')
ax = kmf.plot()

kmf.fit(T2, C2,label='II')
ax = kmf.plot(ax=ax)

kmf.fit(T3, C3,label='III')
ax = kmf.plot(ax=ax)

kmf.fit(T4, C4,label='IV')
ax = kmf.plot(ax=ax)
```

### ExtraTreesClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
forest = ExtraTreesClassifier(n_estimators=10,random_state=0)
forest.fit(X_train_f, y_train_f)
```

#### feature selection

```python
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```
