```python
def gini(list_of_values):
    sorted_list = sorted(list_of_values)
    height, area = 0, 0
    for value in sorted_list:
        height += value
        area += height - value / 2.
        #print (height, area)
    fair_area = height * len(list_of_values) / 2.
    #print (fair_area)
    return (fair_area - area) / fair_area
```

### DATA (frequency of CDR3 TRB per sample)

```python
#make list of vectors with frequences (pandas)
patients = set()
number_of_reads = set()
cl2 = []
d = defaultdict(list)


with open("melanoma.txt", "r") as f:
    for line in f.readlines():
        if line.find("patient_id") == -1:
            l = line.split("\t")
            if (line.find("TRB") != -1):
                pat = l[0]
                patients.add(pat)
                read = float(l[3])
                d[pat].append(read)
 
                
d2 = defaultdict(list)
for k, v in d.items():
    val = gini(v)
    d2["name"].append(k)
    d2["gini_TRB"].append(val)
    
gini_TRB = pd.DataFrame(d2)
gini_TRB[0:10]  
```
