```python
%matplotlib inline
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import collections
from collections import defaultdict
import seaborn as sns
import itertools
import scipy.cluster.hierarchy as hac
import matplotlib.pyplot as plt
```

### Подготовка данных

```python
pat_list = set()
hla_list = set()
data_dict = defaultdict(list)
d = {}


with open("zlatko_new.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        if l[2].find("TCGA") != -1:
            name = l[2]
            pat_list.add(name)
            HLA = l[1].split(',')
            for el in HLA:
                #el1 = el.split("*")[0]+"*"+el.split("*")[1][0:2]
                hla_list.add(el)
                key = (el, name)
                if key in d.keys():
                    d[key] += 1
                else:
                    d[key] = 1
                
data = []
for el in hla_list:
    bla = []
    for k in pat_list:
        if (el,k) not in d.keys():
            bla.append(0)
        else:
            bla.append(d[(el,k)])
    data.append(bla)
    
#data matrix to DataFrame 
p2 = defaultdict(list)
for (a,b) in zip(data, hla_list):
    p2[b] = a

df2 = pd.DataFrame(p2, index = pat_list)
df2[0:10]
```

### проверка значений

```python
some = defaultdict(list)
with open("zlatko_new.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        if l[2].find("TCGA") != -1:
            name = l[2]
            pat_list.add(name)
            HLA = l[1].split(',')
            for el in HLA:
                some["names"].append(name)
                some["HLA"].append(el)
                
res = pd.DataFrame(some)
res[0:10]
res[(res['names'] == 'TCGA-EE-A3AA') &  (res['HLA'] == 'HLA-A*11')].count()                   
```

### Нормировка данных

```python
df3 = df2.div(df2.sum(axis=1), axis=0)
df3['key1'] = pd.Series(list(pat_list), index=df3.index)
#df4 = df3.dropna(thresh=2)
df3[0:3]
```

### Вычисляем расстояния между парами (для sparse используем jaccard-метрику)

```python
from scipy.spatial.distance import squareform, pdist, jaccard
distance_matrix_jaccard = pd.DataFrame(1-(squareform(pdist(df3.ix[:, 1:-1]), 'jaccard')), columns=df3.key1.unique(), index=df3.key1.unique())
distance_matrix_jaccard[0:3]
```

### Иерархическая кластеризация

```python
from scipy.cluster.hierarchy import *
data_linkage_ward = linkage(distance_matrix_jaccard, method='ward') #matrix
```

### Дендрограмма

```python
titles = distance_matrix_jaccard.index.tolist()
fig, ax = plt.subplots(figsize=(15, 20)) # set size
ax = dendrogram(data_linkage_ward, orientation="right", labels =  titles);

plt.tick_params(\
    axis= 'x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off')

plt.tight_layout() #show plot with tight layout
```

### берем 3 кластера 

```python
assignments = fcluster(data_linkage_ward, 3,'maxclust')
cluster_output = pd.DataFrame({'name':titles , 'cluster':assignments})
cluster_output[0:20]
#cluster_output.to_csv("HLA_to_TREE.txt", sep="\t")
```
