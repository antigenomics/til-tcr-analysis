### Готовим данные: классы пациентов по V-usage, HLA, neoantigenes и survival

```python
#open files with features
HLA = pd.read_csv('HLA_TO_TREE.txt', sep="\t", header = None)
HLA.columns = ["id", "HLA", "name"]

NEOANTI = pd.read_csv('neoanti_TO_TREE.txt', sep="\t", header = None)
NEOANTI.columns = ["id1", "neoanti", "name"]

V_USAGE = pd.read_csv('V_USAGE_TO_TREE.txt', sep="\t", header = None)
V_USAGE.columns = ["id2", "v_usage", "name"]

#merge and drop
new = V_USAGE.merge(HLA, on="name", how="inner").merge(NEOANTI, on="name", how="inner")
new[0:10]

input1 = new.drop(new.columns[[0, 3, 5]], axis=1)
input1[0:5]

#open TCGA to SKCM ids
ids = pd.read_csv("TCGA_to_SKCM.csv", sep='\t')
ids[0:5]

#survival data
survival = pd.read_csv("survival.txt", sep='\t', header= None)
survival.columns = ["name", "surv", "1", "2", "3", "4"]
survival[0:5]
surv_data = survival.merge(ids, on="name", how="inner")
surv_data[0:5]
```

```python
#exclude None from survival column and change type to numeric
s = surv_data.loc[surv_data['surv']!= "None"]
s[['surv', '2']] = s[['surv', '2']].apply(pd.to_numeric)

#define first class as > median
cl = pd.DataFrame(s[s['surv'] > 1098.0].name1)
len(cl)
cl["class"] = [1 for x in range(0,222)]

#define second class as < median
cl2 = pd.DataFrame(s[s['surv'] <= 1098.0].name1)
len(cl2)
cl2["class"] = [2 for x in range(0,216)]

#concatenate tables
classes_out = pd.concat([cl, cl2])
classes_out

inp = input1.rename(index=str, columns={"name": "name1"})
inp[0:5]

input_all = inp.merge(classes_out, on="name1", how="inner")
input_all[0:10]
```

```python
from sklearn import tree
titles = input_all['name1'].tolist()
inp_n = input_all.drop(input_all.columns[[1, 4]], axis=1)
X = inp_n.as_matrix()  #v_usage hla neoanti
Y = np.array(input_all["class"].tolist())
```

### Разобьем равномерно по классам нашу выборку на training и test

#### Stratified K-Folds cross-validator

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=2, shuffle=True)
skf.get_n_splits(X, Y)

for train_index, test_index in skf.split(X, Y):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = Y[train_index], Y[test_index]
```

```python
X_train
```

```python
y_train
```

### Обучим на training set

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

from IPython.display import Image
import pydotplus
dot_data = tree.export_graphviz(clf, out_file=None,     
                         filled=True, rounded=True, class_names = ["1", "2"], 
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  
```

### Проверим на тестовой

### Применим зачем-то на весь набор данных

```python
Y_panda = pd.DataFrame(Y, columns=["real"])
X_panda = pd.DataFrame(clf.predict(X), columns=["predict"])

result1 = pd.concat([Y_panda, X_panda], axis=1)
result2 = pd.concat([input_all, result1], axis=1)
result2[0:10]
```
