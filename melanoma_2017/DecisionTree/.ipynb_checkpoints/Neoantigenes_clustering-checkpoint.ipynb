### переделаем zlatko2.txt для TCGA-id

#### оказалось, что неоантигены есть не для всех пациентов, для TCGA-id есть 341, если потянуть  из SKCM-id в соответствии с TCGA, то нагребется 365 (а всего 447 есть с TRB пациентов)

```python
ids = pd.read_csv("TCGA_to_SKCM.csv", sep='\t')
ids.drop('Unnamed: 0', axis=1, inplace=True) 
ids[0:5]
#ids2 = ids.set_index('name1').to_dict()
#ids2 = ids2['name']
#ids2
```

```python
#patientBarcode	disease	gene	peptide	HLA-alleles
dict_SKCM_peptide = defaultdict(list)
dict_TCGA_peptide = defaultdict(list)
out_zlatko = open("zlatko3.txt", "w")
with open("zlatko2.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        if l[0].find("TCGA") == -1:
            #n = l[0]
            n = "SKCM"+(l[0].split('Pat')[1]).rstrip("0").zfill(3)
            peptide = l[3]
            HLA = l[4].split(',') #list of HLA
            HLA1 = set()
            for el in HLA:
                el = el.strip()
                el = el.split('*')[0]+"*"+el.split('*')[1][0:2]
                HLA1.add(el)
            HLA2 = ",".join(list(HLA1))
            dict_SKCM_peptide["name"].append(n)
            dict_SKCM_peptide["peptide"].append(peptide)
            dict_SKCM_peptide["HLA"].append(HLA2)
        else:
            n = l[0]
            peptide = l[3]
            HLA = l[4].split(',') #list of HLA
            HLA1 = set()
            for el in HLA:
                el = el.strip()
                el = el.split('-')[0]+"-"+el.split('-')[1]#+"*"+el.split('*')[1][0:2]
                HLA1.add(el)
            HLA2 = ",".join(list(HLA1))
            dict_TCGA_peptide["name1"].append(n)
            dict_TCGA_peptide["peptide"].append(peptide)
            dict_TCGA_peptide["HLA"].append(HLA2)
            

zlatko = pd.DataFrame(dict_SKCM_peptide)
zlatko2 = zlatko.merge(ids, on="name", how = "inner")          
zlatko2.drop('name', axis=1, inplace=True)


zlatko2[0:10] 
```

### подготовим данные

```python
pat_list = set()
neoant_list = set()
data_dict = defaultdict(list)
d = {}


with open("zlatko_new.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        if l[2].find("TCGA") != -1:
            name = l[2]
            pat_list.add(name)
            anti = l[3]
            neoant_list.add(anti)
            key = (anti, name)
            if key in d.keys():
                d[key] += 1
            else:
                d[key] = 1
                
data = []
for el in neoant_list:
    bla = []
    for k in pat_list:
        if (el,k) not in d.keys():
            bla.append(0)
        else:
            bla.append(d[(el,k)])
    data.append(bla)
    
#data matrix to DataFrame 
p2 = defaultdict(list)
for (a,b) in zip(data, neoant_list):
    p2[b] = a

df2 = pd.DataFrame(p2, index = pat_list)
```

```python
len(pat_list)
```

### нормировка (нужна ли, если все по 1?)

```python
df3 = df2.div(df2.sum(axis=1), axis=0)
df3['key1'] = pd.Series(list(pat_list), index=df3.index)
#df4 = df3.dropna(thresh=2)
df3[0:3]
```

### Сделаем матрицу расстояний, для sparse возьмем jaccard-метрику (и еще из 1 вычтем)

##### #http://dev.bizo.com/2012/01/clustering-of-sparse-data-using-python.html

```python
from scipy.spatial.distance import squareform, pdist, jaccard
distance_matrix_jaccard = pd.DataFrame(1-(squareform(pdist(df3.ix[:, 1:-1]), 'jaccard')), columns=df3.key1.unique(), index=df3.key1.unique())
distance_matrix_jaccard[0:3]
```

### или же без нормировки 

```python
distance_matrix_jaccard2 = pd.DataFrame(1-(squareform(pdist(df3.ix[:, 1:-1]), 'jaccard')), columns=df3.key1.unique(), index=df3.key1.unique())
distance_matrix_jaccard2[0:3]
```

### Кластеризация иерархическая

```python
from scipy.cluster.hierarchy import *
data_linkage_ward = linkage(distance_matrix_jaccard, method='ward') #matrix
```

### Дендрограмма

```python
titles = distance_matrix_jaccard.index.tolist()
fig, ax = plt.subplots(figsize=(15, 20)) # set size
ax = dendrogram(data_linkage_ward, orientation="right", labels =  titles);

plt.tick_params(\
    axis= 'x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off')

plt.tight_layout() #show plot with tight layout
```

```python
len(titles)
```

```python
assignments = fcluster(data_linkage_ward, 3,'maxclust')
cluster_output = pd.DataFrame({'name':titles , 'cluster':assignments})
cluster_output[0:20]
cluster_output.to_csv("neoanti_TO_TREE.txt", sep="\t")
```
