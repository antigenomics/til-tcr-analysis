```python
#open survival.txt
survival = pd.read_csv("survival.txt", sep='\t', header= None)
survival.columns = ["name", "surv", "1", "2", "3", "4"]
survival[0:5]

#open TCGA to SKCM ids
ids = pd.read_csv("TCGA_to_SKCM.csv", sep='\t')
ids[0:5]

#merge survival with ids
surv_data = survival.merge(ids, on="name", how="inner")
surv_data[0:5]

#delete None and <0, make column type numeric
s = surv_data.loc[surv_data['surv']!= "None"]
s[['surv', '2']] = s[['surv', '2']].apply(pd.to_numeric)
s = s.loc[s['surv'] >= 0]

s = s.loc[s['3'] != 'not reported']

def alert(s):
    if (s["3"] == 'Ia') | (s["3"] == 'Ib') | (s["3"] == 'I') | (s["3"] == 'I/II nos'):
        return 'I'
    elif (s["3"] == 'II') | (s["3"] == 'IIb') | (s["3"] == 'IIa') | (s["3"] == 'IIc'):
        return 'II'
    elif (s["3"] == 'III') | (s["3"] == 'IIIb') | (s["3"] == 'IIIa') | (s["3"] == 'IIIc'):
        return 'III'
    elif (s["3"] == 'Iv'):
        return 'IV'
    elif (s["3"] == '0'):
        return '0'
    else:
        return '-'
        
s['class_status'] = s.apply(alert, axis=1)
s = s.loc[(s['class_status'] != '-') & (s['class_status'] != '0')]
s.groupby(['class_status']).size()
```

### GINI FOR V-gene

```python
def gini(list_of_values):
    sorted_list = sorted(list_of_values)
    height, area = 0, 0
    for value in sorted_list:
        height += value
        area += height - value / 2.
        #print (height, area)
    fair_area = height * len(list_of_values) / 2.
    #print (fair_area)
    return (fair_area - area) / fair_area
```
