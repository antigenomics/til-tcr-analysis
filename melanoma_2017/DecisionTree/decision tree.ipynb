```python
stage = surv_data[['name1','3']]
stage[0:10]
```

```python
stage_classes = pd.DataFrame(list(set(stage['3'].tolist())))
stage_classes["stage_group"] = [x for x in range(15)]
stage_classes = stage_classes.rename(index=str, columns={0: "3"})
stage_classes
```

```python
surv_data_dead = surv_data[surv_data["1"] == "dead"]
surv_data_dead[0:10]
```

```python
#exclude None from survival column and change type to numeric
s = surv_data.loc[surv_data['surv']!= "None"]
s[['surv', '2']] = s[['surv', '2']].apply(pd.to_numeric)
s = s.loc[s['surv'] >= 0]

#define first class as > median
#dead and alive
cl = pd.DataFrame(s[s['surv'] > 1098.0].name1)
len(cl)
cl["class"] = [1 for x in range(0,222)]
```

```python
#define second class as < median
#take only dead
s1 = surv_data_dead.loc[surv_data_dead['surv']!= "None"]
s1[['surv', '2']] = s1[['surv', '2']].apply(pd.to_numeric)

cl2 = pd.DataFrame(s1[s1['surv'] <= 1098.0].name1)
len(cl2)
cl2["class"] = [2 for x in range(0,97)]
```

```python
#concatenate tables
classes_out = pd.concat([cl, cl2])
classes_out

inp = input1.rename(index=str, columns={"name": "name1"})
inp[0:5]

input_all = inp.merge(classes_out, on="name1", how="inner")
input_all[0:10]

#len(input_all)
```

```python
#take gini (not gini_IGH or gini_TRB)
from sklearn import tree
titles = input_all4['name1'].tolist()
inp_n = input_all4.drop(input_all4.columns[[1, 4, 6, 7, 8]], axis=1)
X = inp_n.as_matrix()  #v_usage hla neoanti
Y = np.array(input_all4["class"].tolist())
```

### Stratified K-Folds cross-validator

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=2, shuffle=True)
skf.get_n_splits(X, Y)

for train_index, test_index in skf.split(X, Y):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = Y[train_index], Y[test_index]
```

### learning on training set

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

from IPython.display import Image
import pydotplus
dot_data = tree.export_graphviz(clf, out_file=None,     
                         filled=True, rounded=True, class_names = ["1", "2"], 
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  
```
