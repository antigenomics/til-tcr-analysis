### Data

```python
GINI = pd.read_csv('gini.txt', sep="\t")
GINI1 = GINI.drop(GINI.columns[[0]], axis=1)
GINI1[0:10]
```

```python
#open files with features
HLA = pd.read_csv('hla_2.txt', sep="\t", header = None)
HLA.columns = ["id", "HLA", "name"]

NEOANTI = pd.read_csv('neoanti_2.txt', sep="\t", header = None)
NEOANTI.columns = ["id1", "neoanti", "name"]

V_USAGE = pd.read_csv('v_usage_2.txt', sep="\t", header = None)
V_USAGE.columns = ["id2", "v_usage", "name"]


#merge and drop
new = V_USAGE.merge(HLA, on="name", how="inner").merge(NEOANTI, on="name", how="inner")
new[0:10]

input1 = new.drop(new.columns[[0, 3, 5]], axis=1)
input1[0:5]

#open TCGA to SKCM ids
ids = pd.read_csv("TCGA_to_SKCM.csv", sep='\t')
ids[0:5]

#survival data
survival = pd.read_csv("survival.txt", sep='\t', header= None)
survival.columns = ["name", "surv", "1", "2", "3", "4"]
survival[0:5]
surv_data = survival.merge(ids, on="name", how="inner")
surv_data[0:5]
```

```python
#exclude None from survival column and change type to numeric
s = surv_data.loc[surv_data['surv']!= "None"]
s[['surv', '2']] = s[['surv', '2']].apply(pd.to_numeric)

#define first class as > median
cl = pd.DataFrame(s[s['surv'] > 1098.0].name1)
len(cl)
cl["class"] = [1 for x in range(0,222)]

#define second class as < median
cl2 = pd.DataFrame(s[s['surv'] <= 1098.0].name1)
len(cl2)
cl2["class"] = [2 for x in range(0,216)]

#concatenate tables
classes_out = pd.concat([cl, cl2])
classes_out

inp = input1.rename(index=str, columns={"name": "name1"})
inp[0:5]

input_all = inp.merge(classes_out, on="name1", how="inner")
input_all[0:10]
```

```python
input_all2 = input_all.merge(GINI1, on="name1", how="inner")
input_all2[0:10]
#len(input_all2)
```

```python
input_all2[['v_usage', 'HLA', 'neoanti']] = input_all2[['v_usage', 'HLA', 'neoanti']].apply(pd.to_numeric)
input_all2[0:10]
```

```python
#take gini (not gini_IGH or gini_TRB)
from sklearn import tree
titles = input_all2['name1'].tolist()
inp_n = input_all2.drop(input_all2.columns[[1, 4, 6, 7]], axis=1)
X1 = inp_n.as_matrix()  #v_usage hla neoanti
Y = np.array(input_all2["class"].tolist())
```

### Stratified K-Folds cross-validator

```python
from sklearn import tree
X2 = [[0, 0.55], [1, 1.13]]
Y2 = [0, 1]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X2, Y2)
clf.predict([[2.1, 2.2]])
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=2, shuffle=True)
skf.get_n_splits(X1, Y)

for train_index, test_index in skf.split(X1, Y):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X1[train_index], X1[test_index]
    y_train, y_test = Y[train_index], Y[test_index]
```
