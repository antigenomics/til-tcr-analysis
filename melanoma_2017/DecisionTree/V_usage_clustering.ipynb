```python
%matplotlib inline
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import collections
from collections import defaultdict
import seaborn as sns
import itertools
import scipy.cluster.hierarchy as hac
import matplotlib.pyplot as plt
```

### prepare V_usage, patiens_id for TCR, IG

```python
patients = set()
variants_v = set()
d = {}

with open("melanoma.txt", "r") as f:
    for line in f.readlines():
        if line.find("patient_id") == -1:
            l = line.split("\t")
            pat = l[0]
            #cdr = l[6]
            patients.add(pat)
            v = l[8].split(",")
            for el in v:
                variants_v.add(el)
                key = (el, pat)
                if key in d.keys():
                    d[key] += 1
                else:
                    d[key] = 1
```

```python
data = []
for el in variants_v:
    bla = []
    for k in patients:
        if (el,k) not in d.keys():
            bla.append(0)
        else:
            bla.append(d[(el,k)])
    data.append(bla)

#data matrix to DataFrame 
p2 = defaultdict(list)
for (a,b) in zip(data, variants_v):
    p2[b] = a

df2 = pd.DataFrame(p2, index = patients)
df2[0:3]
```

### Normalisation data

```python
df3 = df2.div(df2.sum(axis=1), axis=0)
df3['key1'] = pd.Series(list(patients), index=df3.index)
df4 = df3.dropna(thresh=2) #drop null row if exists
df4[0:3]
#len(df4)
```

### Distance jensen-shannon

```python
import math
#from dit.divergences import jensen_shannon_divergence
from scipy.stats import entropy
from numpy.linalg import norm
import numpy as np
```

```python
a = df4.ix[:, 1:-1].as_matrix()
a
```

```python
#jensen-shannon metric
def JSD(P, Q):
    _P = P / norm(P, ord=1)
    _Q = Q / norm(Q, ord=1)
    _M = 0.5 * (_P + _Q)
    return 0.5 * (entropy(_P, _M) + entropy(_Q, _M))

#distance matrix
def dist(a):
    items = a.shape[0] #(466, 279) --> 466
    out = np.ones((items,items)) #
    for j in range(items):
        for k in range(j+1, items):
            out[j, k] = JSD(a[j], a[k])
            out[k, j] = out[j, k]
    return out

b = dist(a)
b
```

```python
titles = df4.index.tolist()
distance_JS = pd.DataFrame(b, index = titles)

column_indices =  [x for x in range(466)]
old_names = distance_JS.columns[column_indices]
new_names = titles
distance_JS.rename(columns=dict(zip(old_names, new_names)), inplace=True) #replace int indexes to SKCM_id

distance_JS[0:3]
```

### hierarchical ward clusterization

```python
from scipy.cluster.hierarchy import *
data_linkage_ward = linkage(distance_JS, method='ward') #matrix
```

### dendrogram

```python
fig, ax = plt.subplots(figsize=(15, 20)) # set size
ax = dendrogram(data_linkage_ward, orientation="right", labels =  titles)# truncate_mode='lastp', p=12)

plt.tick_params(\
    axis= 'x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off')

plt.tight_layout() #show plot with tight layout

#plt.savefig('ward_clusters1.png', dpi=300)
```

### Gini

```python
def gini(list_of_values):
    sorted_list = sorted(list_of_values)
    height, area = 0, 0
    for value in sorted_list:
        height += value
        area += height - value / 2.
        #print (height, area)
    fair_area = height * len(list_of_values) / 2.
    #print (fair_area)
    return (fair_area - area) / fair_area
```
