### metadata for allen_bcr

```python
out = open("allen_BCR_metadata.txt", "w")

d = {}
with open("allen_resp.txt") as f:
    for line in f.readlines():
        l = line.split("\t")
        name = l[0]
        resp = l[4]
        d[name] = resp.strip()   
        
s = []
with open("allen_bcr.txt") as f:
    for line in f.readlines():
        l = line.strip()
        s.append(l)
        
out.write("#file.name"+"\t"+"sample.id"+"\t"+"col.name"+"\n")
        
for key, val in d.items():
    for el in s:
        s1 = el.split(".")
        name = s1[0]
        n = name.split("-")
        name1 = n[3]
        if name1 == key:
            out.write("VanAllen_BCR/VanAllen_23_nov_BCR/"+el+"\t"+name1+"\t"+val+"\n")
            
out.close()
```

```python
#make dict
filenames_response = [];filenames_nonresponse = []

with open("metadata1.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        name = l[0]
        resp = l[2]        
        if name != "file_name":
            if resp == "PD":
                filenames_nonresponse.append(name)
            else:
                filenames_response.append(name)      
                
with open("metadata.txt", "r") as f:
    for line in f.readlines():
        l = line.split("\t")
        name = l[0]
        resp = l[2]        
        if resp == "response":
            if name != "file_name":
                filenames_response.append(name)
        if resp == "nonresponse":
            if name != "file_name":
                filenames_nonresponse.append(name)  

                
panda = collections.defaultdict(list)

for el in filenames_response:
    with open(el, "r") as f:
        for line in f.readlines():
            l = line.split("\t")
            if l[0] != "count":
                if l[12] != "" and l[13] != "" and l[11] != "":
                    if l[12] != "-1":
                        a = float(l[12])
                        b = float(l[13])
                        c = float(l[11])
                        f = float(l[0])
                        e = l[3]
                        v = l[4]
                        d = l[5]
                        j = l[6]
                        el1 = el.split(".")[0]
                        panda["name"].append(el1)
                        panda["insert_size"].append(a)
                        panda["strength"].append(b)
                        panda["ndn_size"].append(c)
                        panda["cohort"].append("response")
                        panda["count"].append(f)
                        panda["aa"].append(e)
                        panda["v"].append(v)
                        panda["d"].append(d)
                        panda["j"].append(j)
                        
for el in filenames_nonresponse:
    with open(el, "r") as f:
        for line in f.readlines():
            l = line.split("\t")
            if l[0] != "count":
                if l[12] != "" and l[13] != "" and l[11] != "":
                    if l[12] != "-1":
                        a = float(l[12])
                        b = float(l[13])
                        c = float(l[11])
                        f = float(l[0])
                        e = l[3]
                        v = l[4]
                        d = l[5]
                        j = l[6]
                        el1 = el.split(".")[0]
                        panda["name"].append(el1)
                        panda["insert_size"].append(a)
                        panda["strength"].append(b)
                        panda["ndn_size"].append(c)
                        panda["cohort"].append("nonresponse")
                        panda["count"].append(f)
                        panda["aa"].append(e)
                        panda["v"].append(v)
                        panda["d"].append(d)
                        panda["j"].append(j)                       
                                                       
```

### CDF for VanAllen+Hugo insert_size data  

```python
resp_insert = resp_insert.sort_values()

#add last (largest) value (get unbiased CDF)
resp_insert[len(resp_insert)] = resp_insert.iloc[-1]

#make norm dist massive fron 0 to 1 with len(ser) steps
cum_dist = np.linspace(0.,1.,len(resp_insert))
resp_insert_cdf = pd.Series(cum_dist, index=resp_insert)

lin1, = plt.plot(resp_insert_cdf, label="response", color="red")


nonresp_insert = nonresp_insert.sort_values()

#add last (largest) value (get unbiased CDF)
nonresp_insert[len(nonresp_insert)] = nonresp_insert.iloc[-1]

#make norm dist massive fron 0 to 1 with len(ser) steps
cum_dist2 = np.linspace(0.,1.,len(nonresp_insert))
nonresp_insert_cdf = pd.Series(cum_dist2, index=nonresp_insert)


lin2, = plt.plot(nonresp_insert_cdf, label="nonresponse", color="green")
plt.title("CDF_insert_Hugo+Allen")
plt.legend(handles=[lin1, lin2], loc = 'lower right')

plt.show()
```

### Heatmap for VanAllen+Hugo insert_size vs strength

```python
#data
table2 = table[["insert_size","cohort", "strength"]]
table3 = table2.loc[table2['cohort'] == "response"]
table4 = table2.loc[table2['cohort'] == "nonresponse"]

ins_resp = table3.insert_size
stre_resp = table3.strength

ins_nonresp = table4.insert_size.tolist()
stre_nonresp = table4.strength.tolist()

def heatmap(ins, stre):
    d = {}
    for i in range(len(ins)):
        insert = ins[i]
        strength = stre[i]
        key = (insert, strength)
        if key in d.keys():
            d[key] += 1
        else:
            d[key] = 0
    sorted(ins);sorted(stre);n = len(stre);m = len(ins);
    ins = list(set(ins)); stre = list(set(stre)); 
    data = [] 
    for el in ins:
        bla = []
        for k in stre:
            if (el,k) not in d.keys():
                bla.append(0)   
            else:
                bla.append(d[(el,k)])
        data.append(bla)
    return data

data = heatmap(ins_resp, stre_resp);data2 = heatmap(ins_nonresp, stre_nonresp)
#heatmap
plt.figure(1, figsize = (10, 8))
plt.subplot(223);sns.heatmap(data, cmap='viridis');plt.title('response', fontsize=18)
plt.subplot(222);sns.heatmap(data2, cmap='viridis');plt.title('nonresponse', fontsize=18)
plt.tight_layout() 
plt.show()
```
