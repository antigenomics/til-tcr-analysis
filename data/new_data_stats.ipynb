```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import seaborn as sns
import itertools
```

```python
#reponse
resp1 = {}
d = {}
with open("hugo_resp.txt") as f:
    for line in f.readlines():
        l = line.split("\t")
        name = l[0]
        resp = l[2]
        d[name] = resp 
        
s = []
with open("hugo_bcr.txt") as f:
    for line in f.readlines():
        l = line.strip()
        s.append(l)
        
for key, val in d.items():
    for el in s:
        s1 = el.split(".")
        name = s1[0]
        if name == key:
            resp1[name] = val.strip()

resp2 = defaultdict(list)
for key, val in resp1.items():
    resp2["name"].append(key)
    resp2["response"].append(val)
resp3 = pd.DataFrame(resp2) 
```

```python
#function

def RX(meta, receptor):
    hugoTRB = defaultdict(list)
    hugo1 = defaultdict(list)
    p = "hugo_{}_metadata.txt".format(meta)
    with open(p, "r") as f:
        for line in f.readlines():
            line1 = line.split("\t")
            if line1[0] != "#file.name":
                l = line1[0].split("/")
                l1 = l[2]
                l3 = l1.split(".")[0]
                l2 = "Hugo_{}/Hugo_23_nov_{}/".format(meta, meta)+l1
                g = open(l2, "r")
                for line in g.readlines():
                    line1 = line.split("\t")
                    if line1[0] != "Clone ID":
                        if line1[8].find(receptor) != -1:
                            hugo1[l3].append(int(line1[1]))

    hugoTRB1 = {}
    for key, val in hugo1.items():
        summ = 0
        for el in val:
            summ += el
        hugoTRB1[key] = summ
    for key, val in hugoTRB1.items():
        h = "{}".format(receptor)
        hugoTRB["name"].append(key)
        hugoTRB[h].append(val)
    hugoTRBpandas = pd.DataFrame(hugoTRB)
    return hugoTRBpandas

TRB = RX("TCR", "TRB");TRA = RX("TCR", "TRA");IGH = RX("BCR", "IGH");IGK = RX("BCR", "IGK");IGL = RX("BCR", "IGL")
```

```python
#merge
AB = TRB.merge(TRA, on = "name", how = "inner").merge(IGH, on = "name", how = "inner").merge(IGL, on = "name", how = "inner").merge(IGK,on = "name", how = "inner").merge(resp3,on = "name", how = "inner")
AB
```

```python
#dict of lists ==> list of lists ==> common list ===> pandas of unique vals
def common(meta):
    seqaa = defaultdict(list)
    p = "hugo_{}_metadata.txt".format(meta)
    with open(p, "r") as f:
        for line in f.readlines():
            line1 = line.split("\t")
            if line1[0] != "#file.name":
                l = line1[0].split("/")
                l1 = l[2]
                l3 = l1.split(".")[0]
                l2 = "Hugo_{}/Hugo_23_nov_{}/".format(meta, meta)+l1
                g = open(l2, "r")
                for line in g.readlines():
                    line1 = line.split("\t")
                    if line1[0] != "Clone ID":
                        seqaa[l3].append(line1[7])
    return seqaa

TCRhugo = common("TCR")
BCRhugo = common("BCR") 

spTCRresp = []
#SRR
#pandas with hugo reponse - resp3
srr = []
num = []
for key, val in TCRhugo.items():
        spTCRresp.append(val)
        srr.append(key)
        num.append(len(val))        

u1 = []
num2 = []
for i in range(len(spTCRresp)):
    el = spTCRresp[i]
    new = spTCRresp[:i] + spTCRresp[(i+1):]
    u = []
    newlist = [item for sublist in new for item in sublist]
    for l in el:
        if l not in newlist:
            u.append(l)
    leng = len(u)
    num2.append(leng)
    u1.append(u)
    
#resp
#u1
unique_TCR_hugo_resp = {}
unique_TCR_hugo_resp["name"] = srr
unique_TCR_hugo_resp["unique_clonotypes"] = u1
unique_TCR_hugo_resp["number_of_all_clonotypes"] = num
unique_TCR_hugo_resp["number_of_unique_clonotypes"] = num2
result = pd.DataFrame(unique_TCR_hugo_resp)
result_res = result.merge(resp3,on = "name", how = "inner")
result_res
```
